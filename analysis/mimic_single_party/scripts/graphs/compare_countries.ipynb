```python
## original files
with open('/Users/xiaokaren/MyPythonCode/ranked_choice_voting/rcv_proposal/analysis/stability/results/scotland_results_top4.json') as file:
    scotland_original_stability = json.load(file)

with open('/Users/xiaokaren/MyPythonCode/ranked_choice_voting/rcv_proposal/analysis/stability/results/america_results_top4.json') as file:
    america_original_stability = json.load(file)

with open('/Users/xiaokaren/MyPythonCode/ranked_choice_voting/rcv_proposal/analysis/stability/old/australia_results_top4.json') as file:
    australia_original_stability = json.load(file)

with open('/Users/xiaokaren/MyPythonCode/ranked_choice_voting/rcv_proposal/analysis/spoiler/scotland.json') as file:
    scotland_original_spoiler = json.load(file)

with open('/Users/xiaokaren/MyPythonCode/ranked_choice_voting/rcv_proposal/analysis/spoiler/america.json') as file:
    america_original_spoiler = json.load(file)

with open('/Users/xiaokaren/MyPythonCode/ranked_choice_voting/rcv_proposal/analysis/spoiler/australia.json') as file:
    australia_original_spoiler = json.load(file)

scotland_original_data_stability = scotland_original_stability["metadata"]["method_counts"]
scotland_original_data_stability = {key: scotland_original_data_stability[key] for key in methods if key in scotland_original_data_stability}
america_original_data_stability = america_original_stability["metadata"]["method_counts"]
america_original_data_stability = {key: america_original_data_stability[key] for key in methods if key in america_original_data_stability}
australia_original_data_stability = australia_original_stability["metadata"]["method_counts"]
australia_original_data_stability = {key: australia_original_data_stability[key] for key in methods if key in australia_original_data_stability}

scotland_original_data_spoiler = scotland_original_spoiler["metadata"]["method_counts"]
scotland_original_data_spoiler = {key: scotland_original_data_spoiler[key] for key in methods if key in scotland_original_data_spoiler}
america_original_data_spoiler = america_original_spoiler["metadata"]["method_counts"]
america_original_data_spoiler = {key: america_original_data_spoiler[key] for key in methods if key in america_original_data_spoiler}
australia_original_data_spoiler = australia_original_spoiler["metadata"]["method_counts"]
for m in methods:
    if m not in australia_original_data_spoiler:
        australia_original_data_spoiler[m] = 0
australia_original_data_spoiler = {key: australia_original_data_spoiler[key] for key in methods if key in australia_original_data_spoiler}


```

```python
def create_stability_graph(ax, data, title):
    n = len(methods)
    r = np.arange(n) 
    width = 0.2

    ax.bar(r + 0*width, list(scotland_original_data_stability.values()), color="midnightblue", width=width, label='original scotland')
    ax.bar(r + 1*width, list(america_original_data_stability.values()), color="maroon", width=width, label='original america')
    ax.bar(r + 2*width, list(australia_original_data_stability.values()), color="orange", width=width, label='original australia')
    ax.bar(r + 3*width, list(data.values()), color="lightsteelblue", width=width, label='condensed')

    ax.set_xlabel("Method")
    ax.set_ylabel("Count")
    ax.set_xticks(r + width/2)
    ax.set_xticklabels(methods, rotation=40)
    ax.set_title(title)
    ax.legend()

# Create a single figure with two subplots side by side
fig, axs = plt.subplots(4, 2, figsize=(18, 20))  # 1 row, 2 columns

# Plot on each axis
create_stability_graph(axs[0,0], borda_data_stability, "Borda Score Stability")
create_stability_graph(axs[1,0], mention_data_stability, "Mention Score Stability")
create_stability_graph(axs[2,0], first_place_data_stability, "First Place Score Stability")
create_stability_graph(axs[0,1], borda_tiebreaker_data_stability, "Borda Tiebreaker Stability")
create_stability_graph(axs[1,1], mention_tiebreaker_data_stability, "Mention Tiebreaker Stability")
create_stability_graph(axs[2,1], first_place_tiebreaker_data_stability, "First Place Tiebreaker Stability")
create_stability_graph(axs[3,0], keep_first_data_stability, "Keep First Stability")
create_stability_graph(axs[3,1], keep_last_data_stability, "Keep Last Stability")

plt.tight_layout()
plt.show()
```
